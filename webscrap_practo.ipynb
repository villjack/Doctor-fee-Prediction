```python
import requests
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
```

```python
name=[]
spec =[]
exp = []
loc = []
city = []
fees =[]
recom = []
feedback =[]
url =[]
```

```python
cities = ['Delhi', 'Mumbai', 'Bangalore']
specialities = ['Dentist', 'Gynecologist%2Fobstetrician', 'pediatrician', 'orthopedist', 'dermatologist', 'physiotherapist', 'infertility%20specialist', 'ophthalmologist', 'urologist', 'psychiatrist', 'cardiologist', 'dietitian%2Fnutritionist', 'gastroenterologist', 'neurologist', 'neurosurgeon', 'pulmonologist', 'bariatric%20surgeon', 'rheumatologist', 'chiropractor']
```

```python
for c in cities:
  for specialty in specialities:
    link = requests.get("https://www.practo.com/search/doctors?results_type=doctor&q=%5B%7B%22word%22%3A%22"+str(specialty)+"%22%2C%22autocompleted%22%3Atrue%2C%22category%22%3A%22subspeciality%22%7D%5D&city="+str(c))
    webpage = link.content
    soup = BeautifulSoup(webpage,'lxml')
    details = soup.find_all('div', class_ = "info-section")


    for i in details:
      name.append(i.find('h2').text.strip())
      try:
        spec.append(i.find('div', class_ = "u-d-flex").text.strip())
      except:
        spec.append(np.nan)
      try:
        exp.append(i.find('div', {'data-qa-id':"doctor_experience"}).text.strip())
      except:
        exp.append(np.nan)
      try:
        loc.append(i.find('span', {'data-qa-id':"practice_locality"}).text.strip())
      except:
        loc.append(np.nan)

      try:
        city.append(i.find("span",class_ = "u-t-capitalize").text.strip())
      except:
        city.append(np.nan)
  
      try:
        fees.append(i.find("span", {'data-qa-id':"consultation_fee"}).text.strip())
      except:
        fees.append(np.nan)
      try:
        recom.append(i.find("span", {'data-qa-id':"doctor_recommendation"}).text.strip())
      except:
        recom.append(np.nan)
      try:
        feedback.append(i.find("span", {'data-qa-id':"total_feedback"}).text.strip())
      except:
        feedback.append(np.nan)
      try:
        url.append(i.find("a")['href'])
      except:
        url.append(np.nan)
```

```python
print(len(name),
len(spec),
len(exp),
len(loc),
len(city),
len(fees),
len(recom),
len(feedback))
len(url)
```

```python
degree=[]
```

```python
links=[]
for u in url:
  try:
    links.append(("https://www.practo.com"+str(u)))
  except:
    links.append(np.nan)
```

```python
for link in links:
  try:
    webpage1= requests.get(link).content
    soup1 = BeautifulSoup(webpage1,'lxml')
    details1 = soup1.find_all('div', class_ = "pure-u-20-24")
    for d in details1:
      degree.append(d.find('p',{'data-qa-id':"doctor-qualifications"}).text.strip())
  except:
    degree.append(np.nan) 
```

```python
print(name)
print(spec)
print(exp)
print(loc)
print(city)
print(fees)
print(recom)
print(feedback)
print(degree)
```

```python
print(len(name),
len(spec),
len(exp),
len(loc),
len(city),
len(fees),
len(recom),
len(feedback),
len(degree))
```

```python
df = pd.DataFrame(list(zip(name,spec,exp,loc,city,fees,recom,feedback,degree)), columns = ['Name','Specialization','Experience','Location','City','Fees','Recommendation','Feedback','Degree'])
```
